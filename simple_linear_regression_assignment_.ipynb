```python
import seaborn as sns
import matplotlib.pyplot as plt # for graph
sns.distplot(data['YearsExperience'])
```

```python
import seaborn as sns
sns.distplot(data['Salary'])
```

```python
import statsmodels.formula.api as smf    #lib
```

```python
model=smf.ols('Salary~YearsExperience',data=data).fit()   #ORDINARY LEAST SQR
```

```python
model.params
```

```python
model.fittedvalues   #predicrted value
```

```python
model.resid  #error value
```

```python
data['predicted value']=model.fittedvalues
data
```

```python
data['error']=model.resid
data
```

```python
model.params
```

```python
data['y']=25792.200199+(9449.962321*data['YearsExperience']+data['error'])    #y=int+(slope*)
data
```

```python
model.rsquared   #to find the accuracy
```

```python
model.rsquared_adj  # adjectv
```

```python
model.pvalues
```

#transformation method

```python
data['sqrt_yearofexp']=np.sqrt(data['YearsExperience'])
data
```

```python
model=smf.ols('Salary~sqrt_yearofexp',data=data).fit()
```

```python
data['sqrt_predvalues']=model.fittedvalues
data
```

```python
data['sqrt_error']=model.resid
data
```

```python
model.params
```

```python
data['sqrt_yearofexp']
```

```python
data['y_sqrt']=-16055.769117+(41500.680583*data['sqrt_yearofexp']+data['sqrt_error'])
data
```

```python
model.rsquared
```

```python
model.pvalues
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(data['Delivery Time'])
```

```python
sns.distplot(data['sorting_time'])
```

```python
data.head()
```

```python
import statsmodels.formula.api as smf
```

```python
data.rename(columns={'Delivery Time':'delivery_time','Sorting Time':'sorting_time'},inplace=True)
data.head()
```

```python
Time=smf.ols('sorting_time~delivery_time', data=data).fit()
```

```python
data['predicted values']=Time.fittedvalues
data['error']=Time.resid
data
```

```python
Time.params
```

```python
data['y']=-0.756673+( 0.413744*data['delivery_time'] + data['error'])
data
```

```python
Time.rsquared
```

```python
Time.pvalues
```
